```python
path = Path("/content/DataNew")
fnames = get_image_files(path)
print(f"Total Images: {len(fnames)}")
def label_func(x): return x.parent.name
dls = ImageDataLoaders.from_path_func(path, fnames, label_func)
```

```python
dls.valid.show_batch(max_n = 7, nrows = 1)
```

```python
num_0 = 0
num_1 = 0
num_2 = 0
num_3 = 0
num_4 = 0
num_5 = 0
num_6 = 0
num_7 = 0
num_8 = 0
num_9 = 0
num_10 = 0

for idx in range(len(dls.train_ds)):
  if dls.train_ds[idx][1].item() == 0: num_0 += 1
  elif dls.train_ds[idx][1].item() == 1: num_1 += 1
  elif dls.train_ds[idx][1].item() == 2: num_2 += 1
  elif dls.train_ds[idx][1].item() == 3: num_3 += 1
  elif dls.train_ds[idx][1].item() == 4: num_4 += 1
  elif dls.train_ds[idx][1].item() == 5: num_5 += 1
  elif dls.train_ds[idx][1].item() == 6: num_6 += 1
  elif dls.train_ds[idx][1].item() == 7: num_7 += 1
  elif dls.train_ds[idx][1].item() == 8: num_8 += 1
  elif dls.train_ds[idx][1].item() == 9: num_9 += 1
  elif dls.train_ds[idx][1].item() == 10: num_10 += 1

print(f"Num 0: {num_0}, Num 1: {num_1}, Num 2: {num_2}, Num 3: {num_3}, Num 4: {num_4}, Num 5: {num_5}, Num 6: {num_6}, Num 7: {num_7}, Num 8: {num_8}, Num 9: {num_9}, Num 10: {num_10}")
print(f"Total: {num_0+num_1+num_2+num_3+num_4+num_5+num_6+num_7+num_8+num_9+num_10}")
```

```python
total = num_0+num_1+num_2+num_3+num_4+num_5+num_6+num_7+num_8+num_9+num_10
w_0 = total / (10 * num_0)
w_1 = total / (10 * num_1)
w_2 = total / (10 * num_2)
w_3 = total / (10 * num_3)
w_4 = total / (10 * num_4)
w_5 = total / (10 * num_5)
w_6 = total / (10 * num_6)
w_7 = total / (10 * num_7)
w_8 = total / (10 * num_8)
w_9 = total / (10 * num_9)
w_10 = total / (10 * num_10)
```

```python
del learn
```

```python
class_weights=torch.FloatTensor([w_0, w_1, w_2, w_3, w_4, w_5, w_6, w_7, w_8, w_9, w_10]).cuda()
```

```python
loss_func = FocalLoss(gamma = 5.0,weight=class_weights)
```

```python
from fastai.vision.all import cnn_learner
```

```python
roc_auc = RocAuc()
```

```python
learn = cnn_learner(dls, resnet18, loss_func=loss_func, metrics=[roc_auc])
```

```python
learn.lr_find(stop_div=False, num_it=200)
```

```python
learn.fine_tune(8, base_lr=0.0010964781)
```

```python
learn.export()
```
